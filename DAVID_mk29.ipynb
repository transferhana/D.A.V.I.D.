```python
# DAVID mk29 코드 예시 (간소화 버전)
import ipywidgets as widgets
from IPython.display import display
from datetime import date

# 위젯들 정의
title_input = widgets.Text(placeholder='할 일 입력', layout=widgets.Layout(width='100%'))
progress_slider = widgets.IntSlider(min=0, max=100, value=0, layout=widgets.Layout(width='100%'))
due_input = widgets.DatePicker(value=date.today(), layout=widgets.Layout(width='100%'))
priority_dropdown = widgets.Dropdown(options=['낮음', '보통', '높음'], layout=widgets.Layout(width='100%'))

add_button = widgets.Button(description='추가', layout=widgets.Layout(width='100%', height='40px'))
save_button = widgets.Button(description='저장', layout=widgets.Layout(width='49.5%', height='35px'), button_style='info')
load_button = widgets.Button(description='불러오기', layout=widgets.Layout(width='49.5%', height='35px'), button_style='warning')

# 출력 레이아웃
display(widgets.HTML('<h3>📂 DAVID mk29</h3>'))
display(title_input)
display(progress_slider)
display(due_input)
display(priority_dropdown)
display(add_button)
display(widgets.HBox([save_button, load_button]))
```
